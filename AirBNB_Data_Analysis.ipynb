```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
data = pd.read_csv("/content/AirBNB dataset.csv",encoding_errors='ignore')
```

```python
#data cleaning
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data['price']
```

```python
#univariate analysis
sns.histplot(data=data,x='price')
```

```python
#identfying outliers
df=data[data['price'] < 1500]

#outliers
sns.boxplot(data=df,x='price')
```

```python
#outliers
sns.histplot(data=df,x='price')
plt.ylabel('frequency')
```

```python
#price distribution
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365',bins=100)
plt.title('availability_365')
plt.ylabel('frequency')
plt.show()
```
